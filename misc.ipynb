```python
def intersect(arr1, arr2):
    dic = {} 
    comb = arr1 + arr2

    for num in comb: 
        if num in dic: 
            dic[num] += 1
        else: 
            dic[num] = 1
    
    inter = []

    for num, val in dic.items(): 
        if val > 1: 
            inter.append(num)
        else: 
            pass
    
    return inter

X = [1,5,9,0]
Y = [3,0,2,9]
print(intersect(X,Y))


    
```

```python
def find_duplicates(lst):
    res = set()
    set1 = set()
    for num in lst: 
        if num in set1:
            res.add(num) 
        else: 
            set1.add(num)
    return res

lst = [1,2,3,1,3,6,5]
print(find_duplicates(lst))
```

```python
def max_product_of_three(arr):

    if len(arr) < 3: 
        raise ValueError("Need at least three numbers")
    
    max1, max2, max3 = float('-inf')
    min1, min2 = float('inf')

    for x in arr: 
        if x > max1: 
            max1, max2, max3 = x, max1, max2
        elif x > max2: 
            max2, max3 = x, max2
        elif x > max3: 
            max3 = x

        if x < min1: 
            min1, min2 = x, min1
        elif x < min2: 
            min2 = x

    return max(max1 * max2 * max3, max1 * min1 * min2)

    


```

```python
def factorial(num): 

    if num < 0: 
        return -1
    if num == 0:
        return 1
    
    if num == 1: 
        return num 
    else: 
        return num * factorial(num-1) 
    


```

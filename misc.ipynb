```python
def intersect(arr1, arr2):
    dic = {} 
    comb = arr1 + arr2

    for num in comb: 
        if num in dic: 
            dic[num] += 1
        else: 
            dic[num] = 1
    
    inter = []

    for num, val in dic.items(): 
        if val > 1: 
            inter.append(num)
        else: 
            pass
    
    return inter

X = [1,5,9,0]
Y = [3,0,2,9]
print(intersect(X,Y))


    
```

```python
def find_duplicates(lst):
    res = set()
    set1 = set()
    for num in lst: 
        if num in set1:
            res.add(num) 
        else: 
            set1.add(num)
    return res

lst = [1,2,3,1,3,6,5]
print(find_duplicates(lst))
```

```python
def max_product_of_three(arr):

    if len(arr) < 3: 
        raise ValueError("Need at least three numbers")
    
    max1, max2, max3 = float('-inf')
    min1, min2 = float('inf')

    for x in arr: 
        if x > max1: 
            max1, max2, max3 = x, max1, max2
        elif x > max2: 
            max2, max3 = x, max2
        elif x > max3: 
            max3 = x

        if x < min1: 
            min1, min2 = x, min1
        elif x < min2: 
            min2 = x

    return max(max1 * max2 * max3, max1 * min1 * min2)

    


```

```python
from collections import defaultdict

def anagram(s:str, t:str): 

    if len(s) != len(t): 
        return False

    dict_s = {}
    dict_t = {}

    for c in s: 
        dict_s[c] = dict_s.get(c, 0) + 1
    for c in t: 
        dict_t[c] = dict_t.get(c, 0) + 1
    
    return dict_s == dict_t
```

```python
import heapq
def topk(nums: list, k: int): 

    if len(nums) == 1:
        return nums

    dic = {}

    for num in nums: 
        dic[num] = dic.get(num, 0) + 1

    heap = []
    
    for num, count in dic.items():
        heapq.heappush(heap, (count, num))
        if len(heap) > k:
            heapq.heappop(heap)
    
    return [num for (count, num) in heap]



nums = [1,1,1,2,2,3]
topk(nums, k = 4)

```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
 
def merge(list1: ListNode, list2: ListNode):

    dummy = ListNode(-1)
    curr = dummy

    while list1 and list2: 
        if list1.val < list2.val: 
            curr.next = list1
            list1 = list1.next
        elif list2.val <= list1.val: 
            curr.next = list2
            list2 = list2.next
        
        curr = curr.next
    
    if list1: 
        curr.next = list1
    if list2: 
        curr.next = list2

    return dummy.next


list1 = [1,2,4]
list2 = [1,3,4]

# merge(list1, list2)

```

```python
def product(nums: list): 

    answer = [1] * len(nums)
    prefix = 1

    for i in range(len(nums)): 
        answer[i] = prefix
        prefix *= nums[i]
        
    
    suffix = 1

    for i in range(len(nums)-1, -1, -1): 
        answer[i] *= suffix
        suffix *= nums[i]
    
    return answer
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def maxheight_recursive(self, root: TreeNode): 

        if not root:
            return 0 
        
        return 1 + max(self.maxHeight_recursive(root.left), self.maxHeight_recursive(root.right)) 
    
    def maxHeight_iterative(self, root: TreeNode): 
        if not root:
            return 0

        res = 0
        stack = [root]
        depths = [1]

        while stack: 
            node = stack.pop()
            depth = depths.pop()
            if node: 
                res = max(res, depth)
                stack.append(node.right); depths.append(depth + 1)
                stack.append(node.left); depths.append(depth + 1)

        return res

```

```python
def subsets_with_duplicates(nums: list[int]) -> list[list[int]]:

    nums.sort()

    res = []

    def dfs(start, path): 
        res.append(path.copy())

        if start == len(nums): 
            return
        
        for i in range(start, len(nums)): 
            if i > start and nums[i] == nums[i-1]:
                continue
            path.append(nums[i])
            dfs(i + 1, path)
            path.pop()
    
    dfs(0, [])
    return res

print(subsets_with_duplicates(nums = [1,2,2]))

```

```python
def num_islands(grid):

    if not grid or not grid[0]: 
        return 0

    rows = len(grid)
    cols = len(grid[0])
    islands = 0

    DIRNS = [(1,0),(-1,0),(0,1),(0,-1)] 

    def dfs(r, c): 
        if (r < 0
            or r >= rows
            or c < 0 
            or c >= cols
            or grid[r][c] == "0"
            ):
            return 
        
        grid[r][c] = "0"

        for dr, dc in DIRNS: 
            nr = r + dr
            nc = r + dc
            dfs(nr, nc)
        
    for r in range(rows): 
        for c in range(cols): 
            if grid[r][c] == "1": 
                islands += 1
                dfs(r,c)
    
    return islands

                    

    
    
```

```python
def twoSum(arr, target): 
    prevMap = {}

    for i, num in enumerate(arr): 
        diff = target - num
        if diff in prevMap: 
            return [prevMap[diff], i]
        prevMap[num] = i
    
    return []
```
